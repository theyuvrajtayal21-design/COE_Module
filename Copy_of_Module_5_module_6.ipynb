```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Titanic-Dataset.csv")

# Histogram
plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Boxplot
sns.boxplot(x=df['Fare'])
plt.title("Fare Distribution")
plt.show()

# Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Scatter
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()

# Bar chart
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("/content/Titanic-Dataset.csv")

# T-test
survived = df[df['Survived'] == 1]['Age'].dropna()
not_survived = df[df['Survived'] == 0]['Age'].dropna()

t_stat, p_val = stats.ttest_ind(survived, not_survived)

print("T-test")
print("T-statistic:", t_stat)
print("P-value:", p_val)

# Chi-square test
contingency = pd.crosstab(df['Sex'], df['Survived'])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("\nChi-square Test")
print("Chi2:", chi2)
print("P-value:", p)
```

```python
def is_armstrong(n):
    power = len(str(n))
    return n == sum(int(d) ** power for d in str(n))

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def is_perfect(n):
    if n < 1:
        return False
    return n == sum(i for i in range(1, n) if n % i == 0)

num = int(input("Enter a number: "))

print("Armstrong:", is_armstrong(num))
print("Prime:", is_prime(num))
print("Perfect:", is_perfect(num))
```
