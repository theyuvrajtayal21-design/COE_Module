## Task 3: Indexing, Filtering & Sorting (1 Mark)

```python
import pandas as pd

#REDEFINING THE df
data = {
    'Name' : ['Aman','Riya','Rahul','Sneha','Karan'],
   'Department':['IT','HR','IT','Finance','HR'],
    'Salary': [50000,60000,55000,65000,52000]
}
df = pd.DataFrame(data)

# Perform:
# 1. Extract first 3 rows using .loc[]
print("First 3 rows using loc: ")
print(df.loc[0:2])

# 2. Extract Salary column using .iloc[]
print("\nSalary column using iloc:")
print(df.iloc[:, 2]) # Salary is the 3rd column (index 2)

# 3. Filter employees with Salary > 55000
print("\nEmployees with Salary > 55000:")
print(df[df['Salary'] > 55000])

# 4. Sort dataset by Salary (descending)
print("\nDataset sorted by Salary (descending):")
print(df.sort_values(by='Salary', ascending=False))
```

## Task 4: Group-by & Aggregation (1 Mark)

```python
# Perform:
# 1. Find average salary department-wise
print("\nAverage Salary per Department:")
print(df.groupby('Department')['Salary'].mean())

# 2. Find total and count of employees per department
print("\nTotal Salary per Department:")
print(df.groupby('Department')['Salary'].sum())

print("\nEmployee Count per Department:")
print(df.groupby('Department').size())
```

## Task 5: Data Cleaning & Time-Series (1 Mark)

```python
# Given:
df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']

# Perform:
# 1. Convert Joining_Date to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
print("\nDataFrame with Joining_Date converted to datetime:")
print(df.info())

# 2. Extract Year and Month
df['Joining_Year'] = df['Joining_Date'].dt.year
df['Joining_Month'] = df['Joining_Date'].dt.month
print("\nDataFrame with Year and Month extracted:")
print(df[['Joining_Date', 'Joining_Year', 'Joining_Month']])

# 3. Identify duplicate rows (create a temporary duplicate row for demonstration)
df_temp = df.copy()
duplicate_row = {'Name': 'Aman', 'Department': 'IT', 'Salary': 50000, 'Joining_Date': '2022-01-10', 'Joining_Year': 2022, 'Joining_Month': 1}
df_temp.loc[len(df_temp)] = duplicate_row
print("\nDuplicate rows identified (demonstration with a temporary duplicate):")
print(df_temp[df_temp.duplicated()])

# 4. Remove duplicate rows (demonstration with the temporary duplicate)
df_cleaned = df_temp.drop_duplicates()
print("\nDataFrame after Remove Duplicate rows:")
print(df_cleaned)
```
